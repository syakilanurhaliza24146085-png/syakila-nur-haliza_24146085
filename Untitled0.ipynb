```python
import random
import matplotlib.pyplot as plt

# ==========================================
# 1. DATA INPUT (Sesuai Instruksi Soal)
# ==========================================
courses = [
    ("AI", "DosenA"), ("ML", "DosenB"), ("DB", "DosenC"),
    ("CN", "DosenD"), ("SE", "DosenE"), ("OS", "DosenF"),
    ("HCI", "DosenE"), ("IR", "DosenA"), ("NLP", "DosenB"),
    ("CV", "DosenC"), ("IOT", "DosenD"), ("DL", "DosenA"),
    ("CyberSec", "DosenF"), ("BigData", "DosenA"), ("NOSQL", "DosenC"),
    ("Cloud", "DosenD"), ("UX", "DosenE"), ("RL", "DosenB"),
    ("ETL", "DosenB"), ("GameDev", "DosenE"), ("Blockchain", "DosenC"),
    ("SecEng", "DosenF"), ("DevOps", "DosenD"), ("Agile", "DosenF")
]

rooms = ["R1", "R2", "R3", "R4"]

timeslots = [
    "Senin-08", "Senin-10", "Senin-13",
    "Selasa-08", "Selasa-10", "Selasa-13",
    "Rabu-08", "Rabu-10", "Rabu-13",
    "Kamis-08", "Kamis-10", "Kamis-13"
]

# ==========================================
# 2. PARAMETER ALGORITMA GENETIKA
# ==========================================
POPULATION_SIZE = 60
MAX_GENERATIONS = 100
CROSSOVER_RATE = 0.85
MUTATION_RATE = 0.20

# ==========================================
# 3. FUNGSI EVALUASI FITNESS & KONFLIK
# ==========================================
def calculate_fitness(individual):
    conflicts = 0
    room_time_slots = set()
    lecturer_time_slots = set()
    lecturer_days = {}

    for i, gene in enumerate(individual):
        mk, dosen = courses[i]
        room = gene['room']
        time = gene['timeslot']
        day = time.split('-')[0]

        # 1. Konflik Ruang
        if (room, time) in room_time_slots:
            conflicts += 1
        else:
            room_time_slots.add((room, time))

        # 2. Konflik Dosen-Waktu
        if (dosen, time) in lecturer_time_slots:
            conflicts += 1
        else:
            lecturer_time_slots.add((dosen, time))

        # 3. Konflik Dosen-Hari[cite: 1]
        if dosen not in lecturer_days:
            lecturer_days[dosen] = []
        if day in lecturer_days[dosen]:
            conflicts += 1
        else:
            lecturer_days[dosen].append(day)

    # Formula fitness skala 0-100[cite: 1]
    fitness = 100 / (1 + conflicts)
    return fitness, conflicts

# ==========================================
# 4. FUNGSI UTAMA ALGORITMA GENETIKA
# ==========================================
def create_individual():
    return [{'room': random.choice(rooms), 'timeslot': random.choice(timeslots)} for _ in courses]

def crossover(parent1, parent2):
    if random.random() < CROSSOVER_RATE:
        point = random.randint(1, len(courses) - 1)
        child1 = parent1[:point] + parent2[point:]
        child2 = parent2[:point] + parent1[point:]
        return child1, child2
    return parent1.copy(), parent2.copy()

def mutate(individual):
    for gene in individual:
        if random.random() < MUTATION_RATE:
            gene['room'] = random.choice(rooms)
            gene['timeslot'] = random.choice(timeslots)

# ==========================================
# 5. PROSES EVOLUSI (MAIN LOOP)
# ==========================================
population = [create_individual() for _ in range(POPULATION_SIZE)]

history_max_fitness = []
history_avg_fitness = []

print("=== PROSES GENERASI ===")
for generation in range(1, MAX_GENERATIONS + 1):
    evaluated_pop = []
    total_fitness = 0

    for ind in population:
        fit, conf = calculate_fitness(ind)
        evaluated_pop.append((ind, fit, conf))
        total_fitness += fit

    evaluated_pop.sort(key=lambda x: x[1], reverse=True)

    max_fitness = evaluated_pop[0][1]
    avg_fitness = total_fitness / POPULATION_SIZE

    history_max_fitness.append(max_fitness)
    history_avg_fitness.append(avg_fitness)

    if generation <= 10:
        print(f"Gen | {generation}: Max Fitness = | {max_fitness:.2f} | Avg Fitness = {avg_fitness:.2f}")

    next_population = [evaluated_pop[0][0], evaluated_pop[1][0]]

    while len(next_population) < POPULATION_SIZE:
        p1 = random.choice(evaluated_pop[:15])[0]
        p2 = random.choice(evaluated_pop[:15])[0]
        c1, c2 = crossover(p1, p2)
        mutate(c1)
        mutate(c2)
        next_population.extend([c1, c2])

    population = next_population[:POPULATION_SIZE]

best_individual, best_fitness, final_conflicts = evaluated_pop[0]

# ==========================================
# 6. OUTPUT HASIL AKHIR (Sesuai Format Soal)
# ==========================================
print("\n=== JADWAL TERBAIK ===")
for i, gene in enumerate(best_individual):
    mk, dosen = courses[i]
    print(f"MK: {mk:<8} | Ruang: {gene['room']} | Waktu: {gene['timeslot']:<9} | Dosen: {dosen}")

print("\n=== HASIL EVALUASI AKHIR ===")
print(f"Fitness Terbaik: {best_fitness:.2f}")
print(f"Jumlah Konflik : {final_conflicts}")

# ==========================================
# 7. VISUALISASI GRAFIK
# ==========================================
plt.figure(figsize=(10, 5))
plt.plot(range(1, MAX_GENERATIONS + 1), history_max_fitness, label="Fitness Maksimum", color="tab:blue", linewidth=2)
plt.plot(range(1, MAX_GENERATIONS + 1), history_avg_fitness, label="Fitness Rata-rata", color="tab:orange", linewidth=2)

plt.title("Perkembangan Fitness - Algoritma Genetika Penjadwalan Mata Kuliah")
plt.xlabel("Generasi")
plt.ylabel("Fitness")
plt.legend()
plt.grid(True)

# Menampilkan grafik di jendela pop-up IDLE
plt.show()
```
