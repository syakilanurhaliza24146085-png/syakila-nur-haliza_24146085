```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Jalur menuju folder dataset di Google Drive kamu
dataset_path = '/content/drive/MyDrive/flowers'
categories = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']

X = []
y = []
img_size = 64

print("Sedang membaca dan memproses citra...")
for category in categories:
    path = os.path.join(dataset_path, category)
    class_num = categories.index(category)

    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            resized_array = cv2.resize(img_array, (img_size, img_size))
            flattened = resized_array.flatten()
            X.append(flattened)
            y.append(class_num)
        except Exception as e:
            pass

X = np.array(X) / 255.0
y = np.array(y)
print(f"Selesai! Total data sukses diproses: {len(X)} gambar.\n")

# PERBAIKAN: Menambahkan stratify=y agar semua kelas bunga masuk ke data training & testing
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=24146085,
    stratify=y
)

print(f"Jumlah Data Training : {len(X_train)}")
print(f"Jumlah Data Testing  : {len(X_test)}")
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

print("Memulai proses training model MLP (mohon tunggu beberapa saat)...")
mlp = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    max_iter=300,
    random_state=24146085, # Tetap menggunakan NIM kamu[cite: 1]
    verbose=True
)
mlp.fit(X_train, y_train)
print("Training model selesai!\n")

# Melakukan prediksi pada data testing
y_pred = mlp.predict(X_test)

# SOLUSI: Mengambil nama kelas yang benar-benar ada di data testing saja agar tidak error
unique_classes_in_data = np.unique(np.concatenate((y_test, y_pred)))
active_categories = [categories[i] for i in unique_classes_in_data]

# Menampilkan classification report dengan presisi 4 digit sesuai soal[cite: 1]
print("\n" + "="*30)
print("     CLASSIFICATION REPORT     ")
print("="*30)
print(classification_report(y_test, y_pred, target_names=active_categories, digits=4))

# Visualisasi Hasil Prediksi Acak (Menyesuaikan kelas yang ada)
print("\nMenampilkan citra sampel beserta hasil prediksi...")
plt.figure(figsize=(15, 15))
plot_idx = 1

for class_num in unique_classes_in_data:
    category = categories[class_num]
    indices = np.where(y_test == class_num)[0]

    # Ambil 5 atau maksimal jumlah data yang tersedia di kelas tersebut
    num_samples = min(5, len(indices))
    if num_samples == 0:
        continue

    selected_indices = random.sample(list(indices), num_samples)

    for idx in selected_indices:
        plt.subplot(5, 5, plot_idx)
        plt.imshow(X_test[idx].reshape(img_size, img_size), cmap='gray')

        true_label = categories[y_test[idx]]
        pred_label = categories[y_pred[idx]]
        color = 'green' if true_label == pred_label else 'red'

        plt.title(f"True: {true_label}\nPred: {pred_label}", color=color, fontsize=10)
        plt.axis('off')
        plot_idx += 1

plt.tight_layout()
plt.show()
```
